Scrapping - Instamart

```python
#%pip install google-play-scraper
from google_play_scraper import Sort, reviews
app_id='in.swiggy.android.instamart'
result=reviews(app_id, count=2, lang='en', country='in', sort=Sort.NEWEST)
result_a,conti_tok=result
reviews_all=[]
ratings_all=[]

while conti_tok and len(reviews_all)<1000:
    result=reviews(app_id, count=200 , continuation_token=conti_tok)
    result_b,conti_tok=result
    for i in result_b:
        reviews_all.append(i['content'])
        ratings_all.append(i['score'])
# print(len(reviews_all))
# print(len(ratings_all))
#print(reviews_all)

import pandas as pd
df=pd.read_csv('Apps_Reviews.csv')
df['Review Text'] = reviews_all
df['Star Rating'] = ratings_all
df.to_csv('Instamart_Reviews.csv', index=False)
```

```python
import pandas as pd
path='Zepto_Reviews.csv'
df=pd.read_csv(path)
print(df.head())
```

```python
import warnings
warnings.filterwarnings('ignore')

#convert all reviews to string
for i in range(len(df['Review Text'])):
    df['Review Text'][i]=str(df['Review Text'][i])

#1. Lower casing
for i in range(len(df['Review Text'])):
    df['Review Text'][i]=df['Review Text'][i].lower()
     
#2. Remove digits
import re
def remove_digits(text):
    result = re.sub(r'\d+', '', text)
    return result
for i in range(len(df['Review Text'])):
    df['Review Text'][i]=remove_digits(df['Review Text'][i])

#3. Remove punctuations, special characters and emojis
def remove_punctuations(text):
    result = re.sub(r'[^\w\s]', '', text)
    return result
for i in range(len(df['Review Text'])):
    df['Review Text'][i]=remove_punctuations(df['Review Text'][i])

#4. Remove extra spaces
def remove_extra_spaces(text):
    result = re.sub(' +', ' ', text)
    return result
for i in range(len(df['Review Text'])):
    df['Review Text'][i]=remove_extra_spaces(df['Review Text'][i])
    #print(df.loc[i,'Review Text'])

#5. Tokenize
from nltk.tokenize import word_tokenize
for i in range(len(df['Review Text'])):
    #df.loc[i,'Review Text']=word_tokenize(df['Review Text'][i])
    df['Review Text'][i]=word_tokenize(df['Review Text'][i])

#6. Remove stopwords
from nltk.corpus import stopwords
words=stopwords.words('english')
for i in range(len(df['Review Text'])):
    #df.loc[i,'Review Text']=[word for word in df.loc[i,'Review Text'] if word not in words]
    df['Review Text'][i]=[word for word in df['Review Text'][i] if word not in words]

#7. Parts of Speech Tagging
from nltk import pos_tag
for i in range(len(df['Review Text'])):
    #df.loc[i,'Review Text']=pos_tag(df.loc[i,'Review Text'])
    df['Review Text'][i]=pos_tag(df['Review Text'][i])

def calc_tag(t): #t is a parameter recieved from lemmitizer- mapping fuction
    if t.startswith('J'):
        return 'a'  #adjective
    elif t.startswith('V'):
        return 'v'  #verb
    elif t.startswith('N'):
        return 'n'  #noun
    elif t.startswith('R'):
        return 'r'  #adverb
    else:
        return 'n'  #noun as default
    
#8. Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
for i in range(len(df['Review Text'])):
    #df.loc[i,'Review Text']=[lemmatizer.lemmatize(word,calc_tag(tag)) for (word,tag) in df.loc[i,'Review Text']]
    df['Review Text'][i]=[lemmatizer.lemmatize(word,calc_tag(tag)) for (word,tag) in df['Review Text'][i]]
    print(df['Review Text'][i])
```

Feature Extraction

```python
#9. N-grams (Bi-grams)
from gensim.models import Phrases
from gensim.models.phrases import Phraser
phrases=Phrases(df['Review Text'],min_count=5,threshold=10)
bigram=Phraser(phrases)
for i in range(len(df['Review Text'])):
    df['Review Text'][i]=bigram[df['Review Text'][i]]
    print(df['Review Text'][i])

#10. BOW
from gensim.corpora import Dictionary
dictionary=Dictionary(df['Review Text'])
print(dictionary.token2id)
dict_bag=[dictionary.doc2bow(text) for text in df['Review Text']]
print(dict_bag)
```

```python
#11. LDA Topic Modeling
from gensim.models.ldamodel import LdaModel
lda_model=LdaModel(corpus=dict_bag,num_topics=4,id2word=dictionary,passes=10)
for idx,topic in lda_model.print_topics(num_words=8):
    print('Topic: {} \nWords: {}'.format(idx,topic))
```

```python
#12. Coherence Score
from gensim.models.coherencemodel import CoherenceModel
coherence_model_lda=CoherenceModel(model=lda_model,texts=df['Review Text'],dictionary=dictionary,coherence='c_v')
coherence_lda=coherence_model_lda.get_coherence()
print('\nCoherence Score: ',coherence_lda)
```
